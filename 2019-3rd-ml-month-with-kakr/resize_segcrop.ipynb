```python
df_train = pd.read_csv('train.csv')
df_test  = pd.read_csv('test.csv')
```

```python
DATA_PATH = './'
os.listdir(DATA_PATH)
```

```python
TRAIN_IMG_PATH = os.path.join(DATA_PATH, 'train_segcrop')
TEST_IMG_PATH = os.path.join(DATA_PATH, 'test_segcrop')
```

```python
df_train = pd.read_csv(os.path.join(DATA_PATH, 'train.csv'))
df_test = pd.read_csv(os.path.join(DATA_PATH, 'test.csv'))
df_class = pd.read_csv(os.path.join(DATA_PATH, 'class.csv'))
```

```python
IMAGE_SIZE = 299

def resizing_img(img_name, size=(IMAGE_SIZE,IMAGE_SIZE)):
    if img_name.split('_')[0] == 'train':
        PATH = TRAIN_IMG_PATH
        data = df_train
    else:
        PATH = TEST_IMG_PATH
        data = df_test

    img = Image.open(os.path.join(PATH, img_name))

    return img.resize(size)
```

```python
TRAIN_SEGCROPPED_PATH = '../segcropped_train'
TEST_SEGCROPPED_PATH = '../segcropped_test'
```

```python
if (os.path.isdir(TRAIN_SEGCROPPED_PATH) == False):
    os.mkdir(TRAIN_SEGCROPPED_PATH)

if (os.path.isdir(TEST_SEGCROPPED_PATH) == False):
    os.mkdir(TEST_SEGCROPPED_PATH)

for i, row in df_train.iterrows():
    SEGCROPPED = resizing_img(row['img_file'])
    SEGCROPPED.save(os.path.join(TRAIN_SEGCROPPED_PATH, row['img_file']))

for i, row in df_test.iterrows():
    SEGCROPPED = resizing_img(row['img_file'])
    SEGCROPPED.save(os.path.join(TEST_SEGCROPPED_PATH, row['img_file']))
```
