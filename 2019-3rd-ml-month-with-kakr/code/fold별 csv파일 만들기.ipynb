CV를 위해 KFold를 생성한다.
전체 데이터에서의 class 간 비율을 fold에서도 동일하게 유지하기 위해 StratifiedKFold를 사용한다.

```python
from sklearn.model_selection import StratifiedKFold, KFold
skfold = StratifiedKFold(n_splits = 5, random_state = 1993, shuffle = True)
```

모델을 Train 합니다.
StratifiedKFold은 label의 구성비율을 유지하면서 fold를 만들어내므로 split() 호출 시, y(label)도 넘겨야 합니다.

```python
import ssl
from keras.models import model_from_json
from sklearn.utils import shuffle

for fold_index, (train_idx, val_idx) in enumerate(skfold.split(X=df_train['img_file'], y = df_train['class'])):
    # train / validation 데이터 나누기
    dfdf_train = df_train.iloc[train_idx, :].reset_index()
    dfdf_val = df_train.iloc[val_idx, :].reset_index()
    
    # 아래 안해주면 에러 발생. categorical이어서 기준 col이 숫자값이면 안되는 것.
    dfdf_train['class'] = dfdf_train['class'].astype('str')
    dfdf_val['class'] = dfdf_val['class'].astype('str')
    
    df_train_filename = ("fold_%d_train.csv" % (fold_index))
    df_val_filename = ("fold_%d_val.csv" % (fold_index))
    
    dfdf_train = shuffle(dfdf_train).reset_index(drop=True)
    dfdf_val = shuffle(dfdf_val).reset_index(drop=True)
    
    dfdf_train.to_csv(df_train_filename, index=False)
    dfdf_val.to_csv(df_val_filename, index=False)
    
    dft = pd.read_csv(df_train_filename)
    dfv = pd.read_csv(df_val_filename)
    
    print("")
    print("")
    print("fold  %d" % fold_index)
    print("=========================================================")
    print("src:")
    print(dfdf_train.head())
    print("file:")
    print(dft.head())

    print("")

    print("src:")
    print(dfdf_val.head())
    print("file:")
    print(dfv.head())
    
    
    
    
    
    
    
    
    
    
    
```
